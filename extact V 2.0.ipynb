```python
# Define the start and end page numbers (inclusive)
start_page = 431
end_page = 494

pdf_path = 'PMBOK6-2017.pdf' # Replace with the path to your PDF file

# Define an absolute path for the output folder
output_folder = '/content/pdf_images' # You can change this to your desired path

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Convert PDF to images for the specified page range
images = convert_from_path(pdf_path, output_folder=output_folder, fmt='jpg', first_page=start_page, last_page=end_page)

# Initialize the OCR reader
reader = easyocr.Reader(['en']) # You can specify other languages if needed

for i, image in enumerate(images):
    page_number = start_page + i # Calculate the page number for the current image

    # Perform OCR on the image
    result = reader.readtext(f'{output_folder}/page_{page_number}.jpg')

    # Extract and print the detected text
    for detection in result:
        text = detection[1]
        print(f'Page {page_number} Text: {text}')
```
